#**Qu'est ce que le scalping ?**

Le scalping ou micro trading est une technique boursière qui consiste à faire des aller-retours sans cesse en bourse pour acheter et vendre afin de profiter de la variation des cours. Le scalping est un type d'opération boursière très rapide, les ordres et positions prises sur le marché sont revendus ou rachetés après quelques secondes ou minutes seulement. Les exemples les plus courants de scalping sont le scalping sur futures et sur CFD .

##**Comment faire du scalping ?**

Le scalping joue sur des variations très faibles du marché, son objectif est de seulement quelques ticks. Pour que le scalping soit rentable il faut alors choisir un broker qui a des frais de commissions très faibles puisque le scalping demande de nombreux allers et retours. Ensuite le scalping demande une réactivité importante et des nerfs d'acier puisque une position bénéficiaire peut rapidement perdre quelques ticks et devenir déficitaire. Enfin pour faire du scalping il faut être disponible et surveiller ses positions très régulièrement.


##**Les avantages et inconvénients du micro trading**

Le premier avantage est que les gains sont rapide donc motivant et les pertes sont minimisées avec une bonne utilisation des ordres stop. Cependant le scalping ne permet que des gains limités puisqu'il joue sur une fluctuation minime du marché, il faudra donc prendre un nombre important d'opérations par jour pour espérer des gains conséquents. Il est aussi possible de scalper avec un capital ou un levier important (et donc prendre plus de lots) pour augmenter ses gains en scalping, attention cependant, un levier important signifie également un risque accru de perte du capital investi.


##**Le trading à haute fréquence**

Le trading à haute fréquence THF ou en terme anglo-saxon HFT ( High-Frequency Trading) est lui aussi une forme de scalping. Cependant dans le cas du trading à haute fréquence ce sont des robots (proche de super calculateurs)qui grâce à des algorithmes et une connexion très haut débits proche d'une place boursière prennent des ordres de très courtes durée sur les marchés avec un nombre d'aller et retour dépassant depuis 2010 les 1000 aller-retours à la seconde.

* Auto Trendline:

    * Les lignes de tendance peuvent être utilisées pour identifier et confirmer les tendances. Une ligne de tendance relie au moins deux points de prix sur un graphique et est généralement prolongée vers l'avant pour identifier les zones inclinées de support et de résistance. Les lignes avec une pente positive qui soutiennent l'action des prix montrent que la demande nette augmente. Tant que l'action sur les prix reste au-dessus de cette ligne, nous avons une tendance haussière. Les lignes avec une pente négative qui constituent une résistance à l'action sur les prix montrent que l'offre nette augmente. Tant que le prix reste en dessous de cette ligne, la tendance est baissière.

    * Le prix teste généralement plusieurs fois une ligne de tendance inclinée, jusqu'à ce qu'elle se rompe, et à ce moment-là, nous pouvons avoir un renversement de tendance. Les différentes stratégies ont des règles différentes sur la distance qui peut séparer les points de prix connectés et sur la question de savoir s'il faut connecter des mèches ou des corps de bougie, mais toutes les lignes de tendance finissent par se rompre. * Canal de Regression :
  * Description 
    En statistique, la "droite de régression linéaire" est une méthode qui permet de prévoir les valeurs futures à partir d'une distribution de valeurs passées. Dans le cas de l'analyse technique des valeurs boursières, on utilise communément cette méthode pour déterminer quand les cours sortent de la tendance en cours.
    La Droite de Régression Linéaire est une ligne tracée entre deux points en utilisant la méthode des "moindres carrés". Cette ligne est calculée de telle façon que la somme des écarts quadratiques (la somme des carrés des écarts entre le cours d'un jour et le point de la droite se trouvant sur la même verticale que ce cours) soit minimum.
    Cette droite représente donc la "meilleure approximation moyenne" de la position des cours sur la période considérée.
    Pour obtenir la droite de régression linéaire sur une période donnée, on peut utiliser les cours de clôture ou bien les cours les plus hauts, les plus bas ou médian.
    Classiquement en Analyse Technique, comme introduit par Gilbert Raff, la droite de régression linéaire est utilisée pour déterminer un canal avec deux droites parallèles équidistantes en dessus et au-dessous de la droite de régression linéaire.
    La distance entre l'une des lignes du canal et la droite de régression linéaire est appelé "l'écart type" correspondant à la moyenne des écarts quadratiques des cours par rapport à la droite de régression linéaire sur la période fixée. 

    * Calcul
    Le canal de Régression de Raff est construit en traçant deux parallèles. La distance entre les droites du canal et la droite de régression linéaire correspond à la plus grande distance entre un haut ou un bas des prix et la droite de régression linéaire. Le canal est dessiné à partir du niveau de prix le plus distant de la droite de régression linéaire.
    
    La droite du centre est la droite de régression linéaire. Pour calculer les droites supérieures et inférieures du canal, rechercher d'abord le le haut ou le bas le plus éloigné de la droite de régression linéaire. Ensuite, mesurez la distance entre la droite de régression linéaire et ce haut ou ce bas. En fin, ajoutez cette valeur à la droite de régression linéaire pour obtenir la droite supérieur du canal et soustrayez cette valeur de la droite de régression linéaire pour obtenir la droite inférieure du canal.
    
  #### ***Étant  donné que les droites supérieures et inférieures du canal sont tracées au niveau des prix négociés les plus éloignés de la droite de régression linéaire, il est mathématiquement impossible pour le titre de s'échanger à l'extérieur de ces canaux.***

    Beaucoup de logiciels d'analyse technique ou de plateformes fournies par les courtiers en ligne proposent cet indicateur. Il se révélera utile pour placer des stops optimaux, juste sous le canal haussier pour une vente ou juste au-dessus dans le cas d’une vente à découvert dans une tendance baissière.

  * Analyse : 

    * La droite de régression linéaire montre une évolution "équilibrée" des cours.
    Le canal déterminé à partir de cette droite indique où l'on peut attendre une évolution des cours analogue à la tendance et à la volatilité qui ont prévalues lors de la période retenue.

   * A ce titre, les deux droites parallèles constituant le canal peuvent être considérées comme des droites de support et de résistance.

   * Les cours peuvent sortir de ce canal pendant une courte période. Mais si cette sortie dure, c'est le signe d'un retournement de tendance imminent.

* MACD Zero Lag:

  La MACD Zero Lag (ou encore appelée MACD zéro retard) est une variante de l'indicateur MACD classique. Cette variante a été développée par Patrick Mulloy et présentée dans son livre Stocks & Commodities en 1994.

  Par rapport à la MACD classique, la MACD Zero Lag donne des signaux plus précoces. Cependant comme elle réagit plus rapidement aux changements de cours, le risque d'obtenir des faux signaux est plus important.


  * Méthode de calcul:


      MACD-ZL = (2 * MME(cours de clôture, Ct) - MME(MME(cours de clôture, Ct), Ct))
                - (2 * MME(cours de clôture, Lt) - MME(MME(cours de clôture, Lt), Lt))

      Trigger = (2 * MME(MACD-ZL, Tr) - MME(MME(MACD-ZL, Tr), Tr))

où :

    Ct : la période courte
    Lt : la période longue
    Tr : la période du trigger
    MME(cours de clôture, Ct) : la Moyenne Mobile Exponentielle de période Ct des cours de clôture
    MME(cours de clôture, Lt) : la Moyenne Mobile Exponentielle de période Lt des cours de clôture

 * Chartisme:

  Elle permet de mettre à jour les figures chartistes. Ces dernières présentent la caractéristique d’indiquer au trader la tendance à un instant T. Elles lui permettent alors de prendre position en conséquence : acheter ou vendre ? Investir plus ou solder ? Etc.

  Le principe de cette analyse repose sur le fait que les courbes de prix révèlent souvent les mêmes graphiques. Et à un dessin correspond une tendance particulière au sein du marché.
  Lorsqu’il se livre à l’analyse chartiste, l’investisseur cherche donc à identifier un dessein particulier et de prendre position en conséquence.
  Il existe plusieurs figures chartistes :

    la tête et épaules,
    les triangles symétriques,
    les triangles rectangles,
    les diamants,
    les biseaux, etc.

Il est, de ce fait, particulièrement important pour le trader néophyte de se former aux caractéristiques de ces différentes figures. L’expérience constitue aussi un autre atout permettant la maîtrise de ces dernières.

Il n’est pas toujours évident d’analyser les figures chartistes et d’adopter la position qui convient, en conséquence. Pour le trader néophyte, le risque d’effectuer une analyse erronée est toujours présent et pourrait l’empêcher de se fier aux résultats des figures chartistes. Il existe toutefois quelques astuces pour leur permettre d’amoindrir ce risque.

  * Identifier les facteurs difficilement maîtrisables:

    Un pull back trop important ou encore des objectifs qui ne se réalisent pas constituent des facteurs qui faussent les résultats de l’analyse effectuée par le trader. Un investisseur qui ne possède aucune expérience du marché devrait attendre d’autres signaux pour confirmer sa position.

  Ainsi, dans des situations pouvant donner naissance à un pull back, au lieu de prendre immédiatement position en prévision de l’apparition de cette figure, le trader devrait patienter qu’il se réalise. Cela lui évite alors de devoir essuyer des pertes conséquentes dues à une prise de risque trop importante.

  Une astuce consiste, dans ce cadre, à prendre position juste au moment où le pull back frôle la droite. Les risques sont, de ce fait, particulièrement amoindris. Pour pouvoir réaliser des bénéfices dans cette situation, il faut aussi que le trader s’assure de l’exécution de ses ordres au moment opportun. L’identification de ce moment et la mise à disposition de moyens pour une exécution efficace des ordres se présente ainsi comme des incontournables.


## La patience pour des résultats fiables : 
###**Trader des figures chartistes requiert de la patience : plus l’investisseur trade sur des unités de temps importantes, plus les résultats sont fiables. Ainsi, une analyse de la tendance sur un laps de temps d’au moins trente minutes est fortement conseillée.**
